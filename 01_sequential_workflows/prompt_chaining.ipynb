```python
from langchain_groq import ChatGroq
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Optional
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = ChatGroq(model='qwen/qwen3-32b')
```

```python
# create a state
from typing import TypedDict, Optional

class BlogState(TypedDict):
    title: str
    outline: Optional[str]
    content: Optional[str]
    
```

```python
def create_outline(state: BlogState) -> BlogState:

    title = state['title']

    prompt = f"""
You are an expert blog writer.

Create a detailed blog outline for the following topic:

{title}

Include:
- Introduction
- Main sections
- Subsections
- Conclusion
"""

    outline = llm.invoke(prompt).content

    state['outline'] = outline

    return state
```

```python
def create_blog(state: BlogState) -> BlogState:

    # fetch title and outline
    title = state['title']
    outline = state['outline']

    # call the llm gen content
    prompt = f'write a detailed blog on the title - {title} using the following \n {outline}'
    content = llm.invoke(prompt).content

    # update state 
    state['content'] = content

    return state
```

```python
graph = StateGraph(BlogState)

# nodes
graph.add_node('create_outline', create_outline)
graph.add_node('create_blog', create_blog)

# edges
graph.add_edge(START ,'create_outline')
graph.add_edge('create_outline', 'create_blog')
graph.add_edge('create_blog', END)

# compile
workflow = graph.compile()
```

```python
intial_state = {'title': 'Rise of AI in Pakistan'}

final_state = workflow.invoke(intial_state)
final_state
```
